```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('scores_assault_final.csv', index_col = 0)
df
```

```python
df['score'].rolling(window=400).mean().dropna().plot()
```

```python
plt.figure(figsize=(30,6))
n = 500
plt.plot(df[:n]['score'].reset_index(drop= True), alpha = 0.7)
plt.plot(df[-1*n:]['score'].reset_index(drop=True), alpha = 0.7)
```

```python
df['score'].plot(figsize = ((30,6)))
```

```python
plt.scatter(df['score'], df['train_time'])
plt.xlabel('score')
plt.ylabel('train_time')
plt.show()
```
