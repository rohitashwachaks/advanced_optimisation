```python
import pandas as pd
```

```python
# df = pd.read_csv('scores_assault_final_bkp.csv', index_col=0)[['score','train_time']]
df = pd.read_csv('scores_assault_final.csv')[['score','train_time']]
df
```

```python
from matplotlib import pyplot as plt

plt.scatter(df['score'], df['train_time'])
plt.ylabel('Training Time (s)')
plt.xlabel('Score')
plt.title('Scores vs Training Time')
plt.show()
```

```python
df['score'].plot()
```

```python
window = 600
fig, ax = plt.subplots(2,1, figsize=(20,8), sharex=True)
df['score'].rolling(window= window).mean().dropna().plot(ax = ax[0])
ax[0].set_ylabel('Score')
df['train_time'].rolling(window= window).mean().dropna().plot(ax = ax[1])
ax[1].set_ylabel('Time (s)')
plt.xlabel('#games')
plt.suptitle('Training Metrics')
```
